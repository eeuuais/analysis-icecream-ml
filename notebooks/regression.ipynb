```python
# 이상치 개수 확인

# IQR 계산
Q1 = icecream_sales['판매가격'].quantile(0.25)
Q3 = icecream_sales['판매가격'].quantile(0.75)
IQR = Q3 - Q1

# 하한선과 상한선 계산
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 이상치 확인
outliers = icecream_sales[(icecream_sales['판매가격'] < lower_bound) | (icecream_sales['판매가격'] > upper_bound)]

# 이상치 출력
print("이상치 개수:", len(outliers))
print(outliers)
```

# 데이터 일별로 재정비

```python
# '승인일자' 기준으로 그룹화하여 일별 승인금액 합계 계산
daily_sales = icecream_sales.groupby(icecream_sales['승인일자'].dt.date)['판매가격'].sum().reset_index()

# '승인일자'를 datetime으로 다시 변환 (그룹화 후 다시 '승인일자'가 datetime이므로)
daily_sales['승인일자'] = pd.to_datetime(daily_sales['승인일자'])

# 주말 여부를 나타내는 열 추가 (0 = 월요일, 6 = 일요일)
daily_sales['주말'] = daily_sales['승인일자'].dt.weekday >= 5

# 월(Month) 정보 추가
daily_sales['월'] = daily_sales['승인일자'].dt.month

# 요일(Weekday) 정보 추가 (0 = 월요일, 6 = 일요일)
daily_sales['요일'] = daily_sales['승인일자'].dt.weekday

# 결과 확인
daily_sales.head()
```

```python
# 열 이름 변경: '판매가격' -> '승인금액'
daily_sales = icecream_sales.rename(columns={'판매가격': '일별승인금액'})

# 확인
daily_sales.head()
```
